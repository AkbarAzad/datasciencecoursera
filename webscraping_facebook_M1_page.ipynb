```python
request_url = 'https://mbasic.facebook.com/ufi/reaction/profile/browser/fetch/?limit=134&total_count=134&ft_ent_identifier=10158376352191185'
```

```python
# Import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import string
import nltk
from nltk.corpus import stopwords
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import cv2
import os
from selenium import webdriver
import time
import datetime
import secrets
```

```python
# Launch driver
driver = webdriver.Chrome()
```

```python
driver.get(request_url)
```

```python
# Find username box
xpath1 = '//*[@id="m_login_email"]'
```

```python
# Enter username
driver.find_element_by_xpath(xpath1).send_keys(secrets.username)
```

```python
# Find password box
xpath2 = '//*[@id="login_form"]/ul/li[2]/section/input'
```

```python
# Enter password
driver.find_element_by_xpath(xpath2).send_keys(secrets.password)
```

```python
# Find log-in box
xpath3 = '//*[@id="login_form"]/ul/li[3]/input'
```

```python
# Log-in
driver.find_element_by_xpath(xpath3).click()
```

```python
# Create soup
soup3 = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
soup3
```

```python
# Extract likes
names = soup3.find_all('h3', {'class': 'bj'})
```

```python
# Store names of those who like
names_list = []

for name in names:
    print(name.text)
    names_list.append(name.text)
```

```python
# Store names of those who like in dataframe
names_df = pd.DataFrame({'Name': names_list})
```

```python
# Export names as dataframe into CSV
names_df.to_csv('names_df_facebook_m1_post_28112020.csv', index = False)
```

```python
posts = soup4.find_all('div', {'class': 'gl'})
```

```python
posts
```

```python
posts_list = []

for item in posts:
    #soup_item = item.find_all('div', {'class':'gu'})
    item_text = item.text[2:]
    posts_list.append(item_text)
```

```python
posts_df = pd.DataFrame({'Post':posts_list})
posts_df = posts_df.drop_duplicates().reset_index(drop = True)
```

```python
posts_df.to_csv('posts_df_facebook_m1_post_28112020.csv', index = False)
```

```python
articles = soup4.find_all('article', {'class': 'df gj gk'})
```

```python
import datetime as datetime

articles_list = []

for item in articles:
    post = item.find('div', {'class': 'gl'})
    date = item.find('div', {'class':'cm fs'})
    print(date.text)
    print(post.text)
    article_dict = {'Post': post.text[2:],
                   'Date': date.text,
                   'Current Datetime': datetime.datetime.now().strftime(format = '%Y-%m-%d %H:%M:%S')}
    articles_list.append(article_dict)
```

```python
articles_df = pd.DataFrame(articles_list)
articles_df = articles_df.drop_duplicates().reset_index(drop = True)
```

```python
articles_df.to_csv('articles_df_facebook_m1_post_28112020.csv', index = False)
```
