```python
articles = soup4.find_all('article', {'class': 'df gj gk'})
```

```python
import datetime as datetime

articles_list = []

for item in articles:
    post = item.find('div', {'class': 'gl'})
    date = item.find('div', {'class':'cm fs'})
    post_link = item.find_all('a', {'class': 'he hf'})
    comment = item.find_all('a', {'class': 'hf'})
    print(date.text)
    print(post.text)
    for item2 in post_link:
        print(item2.text)
    for item3 in comment[1:]:
        print(item3.text)
    article_dict = {'Post': post.text[2:],
                    'Reactions': item2.text,
                    'Comments': item3.text,
                   'Date': date.text,
                   'Current Datetime': datetime.datetime.now().strftime(format = '%Y-%m-%d %H:%M:%S')}
    articles_list.append(article_dict)
```

```python
articles_df = pd.DataFrame(articles_list)
articles_df = articles_df.drop_duplicates().reset_index(drop = True)
```

```python
articles_df.to_csv('articles_df_facebook_m1_post_' + datetime.datetime.now().strftime(format = '%Y%m%d_%H%M%S') + '.csv', index = False)
```
